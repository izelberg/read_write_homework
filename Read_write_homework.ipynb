Учтите условия на данные:
содержимое purchase_log.txt помещается в оперативную память компьютера

```python
f1 = open('purchase_log.txt', 'r', encoding='utf-8')
```

```python
purchase = f1.readlines()
```

```python
#Специально по шагам делается: для прочитанного целиком в память файла purchase_log.txt, который имеет вид, 
#где много текстовых записей, которые с виду походят на словарь, но словарем не являются; 
#cначала разбиваем на двойные строки типа "user_id": "afea8d72fc"; затем берем каждый такой элемент и разбиваем его на 
#user_id и afea8d72fc, и берем только второй элемент [1] = afea8d72fc, удаляя лишние пробелы и кавычки. Из второго элемента типа 
#"category": "Электроника" тоже берем второй элемент [1] = Электроника.
#Затем записываем каждую строку в словарь
purchases = {}
for p in purchase:
    a = p.strip('{}\n').split(',') #Срезали у строки лишние символы. Разбили на отдельные части, разделенные запятыми.
    a0 = a[0].split(':')[1].strip(' "') #Разбили 1-ю часть строки по :, со второй подчасти срезали лишние символы. Остался только user_id
    a1 = a[1].split(':')[1].strip(' "')
    #print (a0, a1)
    purchases[a0] = a1 #Записываем в словарь purchases значение а1 по ключу а0
#print(purchases)
```

```python
#Пример из лекции. 1. Импорт библиотеки json. 2. Открываем файл и сразу переводим в словарь через json.loads.
#Далее не используется
import json
```

```python
#Пример из лекции. Далее в ДЗ не используется. Альтернативный вариант перевода в словарь. Но в этом словаре остаются 
#названия user_id и category, которые нам не нужны дальше
i = 0
with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        purchase = json.loads(line)
        
        print(purchase)
        
        i += 1
        if i > 5:
            break
```

```python
#Пример из лекции. Далее в ДЗ не используется. 
with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        purchase = json.loads(line)
        print(purchase['user_id']) #Вот так печатает весь словарь
    #print(purchase)  #Вот так печатает выборочную строку, непонятно какую
```

```python
#Открываем файл visit_log.csv и сразу потом результат записываем в файл funnel.csv
#Берем каждую отдельную строку из файла visit_log.csv, ищем первый элемент каждой строки (line[0] - это user_id)
#в словаре purchases и записываем в новый файл funnel: 
#первый элемент из файла visit_log.csv = line[0] = user_id
#второй элемент из файла visit_log.csv = line[0] = source
#значение из словаря purchases по ключу line[0] = category
with open('visit_log.csv', 'r', encoding='utf-8') as f2:
    with open('funnel.csv', 'w', encoding='utf-8') as f3:
        for line in f2:
            line = line.strip().split(',')
            if line[0] in purchases:
                f3.write (line[0] + ', ' + line[1] + ', ' + purchases[line[0]] + '\n')
```
